```python
import matplotlib.font_manager as fm

font_name = fm.FontProperties(
    fname="/usr/share/fonts/Pretendard/PretendardGOV-Medium.ttf"
).get_name()
plt.rc("font", family=font_name)
```

```python
SAVE_PATH = "../database"

IMPLICIT_METRIC = pd.read_csv(f"{SAVE_PATH}/IMPLICIT_METRIC.csv")
IMPLICIT_METRIC = IMPLICIT_METRIC.sort_values(by="scode")
IMPLICIT_METRIC = IMPLICIT_METRIC.set_index("scode")
IMPLICIT_METRIC.head()
```

```python
MAX_VALUE = max(IMPLICIT_METRIC.max().to_list()) *1.15

plt.rcParams["figure.figsize"] = (39 * 3, 10)
ax1 = IMPLICIT_METRIC.loc[95:200, :].plot.bar()
fig1 = ax1.get_figure()
ax1.set_ylim(0, MAX_VALUE)

plt.rcParams["figure.figsize"] = (43 * 3, 10)
ax2 = IMPLICIT_METRIC.loc[201:300, :].plot.bar()
fig2 = ax2.get_figure()
ax2.set_ylim(0, MAX_VALUE)

plt.rcParams["figure.figsize"] = (16 * 3, 10)
ax3 = IMPLICIT_METRIC.loc[301:400, :].plot.bar()
fig3 = ax3.get_figure()
ax3.set_ylim(0, MAX_VALUE)

plt.rcParams["figure.figsize"] = (14 * 3, 10)
ax4 = IMPLICIT_METRIC.loc[401:500, :].plot.bar()
fig4 = ax4.get_figure()
ax4.set_ylim(0, MAX_VALUE)
```

```python
# Save File
figs = [-1, fig1, fig2, fig3, fig4]
for idx, f in enumerate(figs):
    if f == -1:
        continue
    f.savefig(f"line{idx}_implicit_metric.svg", format="svg")
```
