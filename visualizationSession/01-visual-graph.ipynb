```python
subway_graph = dist_weight_graph()
subway_graph = time_weight_graph()
subway_graph.adj_list[96]
```

```python
# Example adjacency list (replace with your own data)
adjacency_list = subway_graph.adj_list

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges from the adjacency list
for node, neighbors in adjacency_list.items():
    G.add_node(node)
    for neighbor, weight in neighbors:
        G.add_edge(node, neighbor, weight=round(weight))

# Draw the graph
pos = nx.spring_layout(G, seed=4034012)  # You can choose other layouts as well
fig, ax = plt.subplots(figsize=(192 / 8, 108 / 8))

labels = nx.get_edge_attributes(G, "weight")
nx.draw(
    G,
    pos,
    ax=ax,
    with_labels=True,
    node_size=300,
    linewidths=1,
    edgecolors="black",
    node_color="#ffffff",
    alpha=0.4,
    font_size=6,
)
nx.draw_networkx_edge_labels(G, pos, ax=ax, edge_labels=labels, font_size=4)

fig.set_facecolor("#ffffff")
# ax.set_facecolor("#ffffff")

plt.show()
fig.savefig("graph.svg", format="svg")
```
