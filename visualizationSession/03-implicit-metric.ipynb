```python
IMPLICIT_METRIC = pd.read_csv(f"../database/IMPLICIT_METRIC.csv")
IMPLICIT_METRIC = IMPLICIT_METRIC.sort_values(by="scode")
IMPLICIT_METRIC = IMPLICIT_METRIC.set_index("scode")
IMPLICIT_METRIC.head()
```

```python
TRIM_A4_W = 7.48031
TRIM_A4_H = 10.9055
```

```python
MAX_VALUE = max(IMPLICIT_METRIC.max().to_list()) * 1.125
plt.rcParams["font.size"] = 7
for line in [1, 2, 3, 4]:
    l_range = line_num_to_range(line)
    figsize = (len(l_range) * TRIM_A4_H / 43, TRIM_A4_W / 2)
    if line == 3 or line == 4:
        figsize = (len(l_range) * TRIM_A4_W/17, TRIM_A4_W / 2)
    print(figsize)
    plt.rcParams["figure.figsize"] = figsize
    ax = IMPLICIT_METRIC.loc[l_range, :].plot.bar()
    ax.set_ylim(0, MAX_VALUE)
    fig = ax.get_figure()
    fig.savefig(f"line{line}_implicit_metric.svg", format="svg")
```
